```python
import multiprocessing
import numpy as np
```

```python

def printVal(value):
    return ("hello "*(value))
pool_obj = multiprocessing.Pool()
ans = pool_obj.map(printVal,range(0,4))
print(ans)
pool_obj.close()
```

```python
extra_keys = ["extra1", "extra2", "extra3"]
input_keys = ["input1", "input2"]

samples = np.array([np.linspace(0.0, 10, 9) for _ in input_keys]).T
print(samples)
```

```python
def interp_eval(sam,vkey):
    sam1, sam2 = sam
    return sam1*2.0 + sam2
```

```python
print(f"Postprocessing grid with extra keys: {extra_keys}")

def postproc(k):
    return [interp_eval(sam, vkey=k) for sam in samples]

output_samples = []
# for k in extra_keys:
#     output_samples.append(postproc(k))
# output_samples = np.array(output_samples).T

pool_obj = multiprocessing.Pool(processes=10)
ans = pool_obj.map(postproc,extra_keys)
output_samples = np.array([np.array(ans[i]) for i in range(len(extra_keys))]).T
print(np.shape(output_samples))

all_samples = np.hstack([samples, output_samples])
all_keys = list(input_keys) + extra_keys
print("    done")
print("")

print(all_samples)
```
